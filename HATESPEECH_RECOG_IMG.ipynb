```
import pytesseract
from PIL import Image
from transformers import pipeline

# Function to extract text from an image using Tesseract
def extract_text_from_image(image_path):
    """
    Extract text from an image using Tesseract OCR.
    """
    image = Image.open(image_path)
    text = pytesseract.image_to_string(image)
    return text

# Initialize a hate speech detection pipeline using a pre-trained transformer model
hate_speech_detector = pipeline("text-classification", model="Hate-speech-CNERG/dehatebert-mono-english")

# Function to classify extracted text for hate speech
def detect_hate_speech(text):
    """
    Classify the given text to detect hate speech.
    """
    results = hate_speech_detector(text)
    return results

# Analyze an image for hate speech
def analyze_image_for_hate_speech(image_path):
    """
    Extract text from an image and check if it contains hate speech.
    """
    # Step 1: Extract text
    text = extract_text_from_image(image_path)
    print("Extracted Text:", text)

    # Step 2: Detect hate speech in the text
    result = detect_hate_speech(text)
    print("Hate Speech Detection Result:", result)
    return result

# Example usage with uploaded image
image_path = list(uploaded.keys())[0]  # Get the uploaded file's name
analyze_image_for_hate_speech(image_path)
```
